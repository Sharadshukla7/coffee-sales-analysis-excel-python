# Total Revenue

```python
total_revenue = df['total_sales'].sum().round(2)
print("Total Revenue : ", total_revenue)
```

# Total Footfall


```python
total_footfall = df['transaction_id'].count()
print("Total Footfall : ", total_footfall)
```

# Average Bill Per Person

```python
average_cust_bill = df['total_sales'].mean()
print("average_cust_bill : ", round(average_cust_bill,2))
```

# Average Order Per Person

```python
average_cust_order = df['transaction_qty'].mean()
print(" Average customer order : ", round(average_cust_order,2)) 
```

# Revenue On Each Day

```python
revenue_on_each_day = df.groupby('Day Name')['total_sales'].sum()
print("Revenue on each day : ", revenue_on_each_day)
```

# Order On Each Day

```python
Order_each_day = df.groupby('Day Name')['transaction_id'].count().reset_index().sort_values(['transaction_id'],ascending = False)
print(" total oprder on each day : ", Order_each_day)
```

# Order And Revenue On Each Day

```python
order_and_revenue_on_each_day = df.groupby('Day Name').agg({'transaction_id': 'count' , 'total_sales': 'sum'})
print("Revenue on each day : ", order_and_revenue_on_each_day)
```

# Revenue per hour 

```python
hour = df.groupby('Hour')['total_sales'].sum()
print("Revenue per hour : ", hour)
```

# Revenue per Month

```python
month = df.groupby('Month Name')['total_sales'].sum().reset_index().sort_values(['total_sales'],ascending = False)
print("Revenue per month : ", month)
```

# Size Distribuation By Order

```python
order_per_size = df.groupby('size')['transaction_qty'].count()
print(" Order per Size : ", 100*(order_per_size/df['transaction_qty'].count()))
```

# Sales and footfall on various locations

```python
sales_footfall_locations = df.groupby('store_location').agg({'transaction_id' : 'count' , 'total_sales' : 'sum'})
print("Footfall and sales over variuous locations : ", sales_footfall_locations)
```

# Revenue Per Category

```python
revenue_category = df.groupby('product_category')['total_sales'].sum().reset_index().sort_values(['total_sales'], ascending = False)
print("Revenue per category : ", revenue_category)
```

# Revenue based top 5 product

```python
revenue_product_type = df.groupby('product_type')['total_sales'].sum().reset_index().sort_values(['total_sales'], ascending = False).head(5)
print("Revenue per product type  : ", revenue_product_type)
```

# Revenue based top 10 product detail

```python
revenue_product_detail = df.groupby('product_detail')['total_sales'].sum().reset_index().sort_values(['total_sales'],ascending = False).head(10)
print(" Revenue Based Top 10 Product Details Are : ", revenue_product_detail)
```
